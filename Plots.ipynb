```python
movies.Genre = movies.Genre.astype('category')
```

```python
genre = []
labels= list()
for gen in movies.Genre.cat.categories:
    genre.append(movies[movies.Genre==gen]['Budget (million $)'])
    labels.append(gen)
plt.hist(genre,stacked=True,bins=30, rwidth=2, label=labels)
plt.legend()
plt.rcParams['figure.figsize']=30,8
plt.show()
```

```python
f, axes = plt.subplots(2,3, figsize=(12,6),sharex=True,sharey=True)
k1 = sb.kdeplot(movies['Budget (million $)'],movies['Audience Ratings %'], ax=axes[0,0])
k2 = sb.kdeplot(movies['Budget (million $)'],movies['Rotten Tomatoes Ratings %'], ax=axes[1,0])
k1 = sb.kdeplot(movies['Budget (million $)'],movies['Audience Ratings %'], ax=axes[0,1])
k2 = sb.kdeplot(movies['Budget (million $)'],movies['Rotten Tomatoes Ratings %'], ax=axes[1,1])
k1 = sb.kdeplot(movies['Budget (million $)'],movies['Audience Ratings %'], ax=axes[0,2])
k2 = sb.kdeplot(movies['Budget (million $)'],movies['Rotten Tomatoes Ratings %'], ax=axes[1,2])
k1.set(xlim=(-20,160))
```
