```python
#from the little you have learnt from variables and research, you should know that CountryName is a independent variable.
# so let's plot it on x-axis and a dependent variable on y-axis. As x is an independent variable, and y depends on how x changes.
import matplotlib.pyplot as mp
```

```python
#mp.plot(dependent variable(y),styles)
mp.plot(InternetUsers[0],c='black',ls='--',marker='s',ms='10',label=InternetUsers[0])
mp.plot(InternetUsers[1],c='black',ls='--',marker='s',ms='10'label=InternetUsers[1])
mp.plot(InternetUsers[2],c='black',ls='--',marker='s',ms='10'label=InternetUsers[2])
mp.plot(InternetUsers[3],c='black',ls='--',marker='s',ms='10'label=InternetUsers[3])
mp.plot(InternetUsers[4],c='black',ls='--',marker='s',ms='10'label=InternetUsers[4])
mp.plot(InternetUsers[5],c='black',ls='--',marker='s',ms='10'label=InternetUsers[5])
mp.xticks(list(range(0,6)),CountryName,order='vertical')
mp.legend(loc='upper left' bbox_to_anchor=(0,1))
```
